```python
next1 =int(input("Enter the number of rows ")); 
for i in range(0,next1):
  for j in range(0,next1-i-1):
    print(end = ' ')
  for j in range(0, i+1):
    print('*',end = ' ')
  print()
```

```python
rows = 10
i = rows
while i >= 1:
    j = rows
    while j > i:
        # display space
        print(' ', end=' ')
        j -= 1
    k = 1
    while k <= i:
        print('*', end=' ')
        k += 1
    print()
    i -= 1
```

```python
rows = 14
print("*" * rows, end="\n")
i = (rows // 2) - 1
j = 2
while i != 0:
    while j <= (rows - 2):
        print("*" * i, end="")
        print(" " * j, end="")
        print("*" * i, end="\n")
        i = i - 1
        j = j + 2
```
